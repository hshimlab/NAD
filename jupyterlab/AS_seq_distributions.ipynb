```python
#Calculate N50
df_Basecall_seqlen_5_sorted = df_Basecall_seqlen_5['seqlen'].sort_values(ascending=False) #sort in descending order
mean = sum(df_Basecall_seqlen_5['seqlen'])

for i in range(len(df_Basecall_seqlen_5_sorted)): #find a contig that contains the mean value
    mean = mean - df_Basecall_seqlen_5_sorted.iloc[i]
    #print(mean/(i+1))
    if mean < (sum(df_Basecall_seqlen_5['seqlen'])/2):
        print(i)
        print(df_Basecall_seqlen_5_sorted.iloc[i-1])
        break
```
